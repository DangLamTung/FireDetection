```python
import glob
import os
import cv2
import numpy as np
from numpy import genfromtxt
import matplotlib.pyplot as plt
```

```python
path = os.getcwd()
#Image path
folderPath = os.path.join(path, 'Captured')
folderList = os.listdir(folderPath)
```

```python
#Sort input by number
folderList = list(map(int, folderList))
folderList.sort()
folderList = list(map(str, folderList))
```

```python
train_data = []
for x in folderList:
    filePath = x
    print(x)
    for i in os.listdir(folderPath + "//" + str(x)):
        subFolderPath = folderPath + "//" + str(x) + "//" + str(i)   
        for filename in glob.glob(os.path.join(subFolderPath, '*.csv')):
            if(filename.find("data")!=-1):
                #print(filename)
                train_data.append(genfromtxt(filename, delimiter=','))
train_data = np.array(train_data)
print(train_data.shape)
```

```python
#Label path
labelPath = os.path.join(path, 'Label')
labelList = os.listdir(labelPath)

label = []

for filename in glob.glob(os.path.join(labelPath, '*.csv')):
    label.append(genfromtxt(filename, delimiter=',')[:,0:290])
label = np.array(label)
print(label.shape)
w=10
h=10
fig=plt.figure(figsize=(10, 10))
columns = 10
rows = 12
for i in range(1, columns*rows-1):
    img = label[i]
    fig.add_subplot(rows, columns, i)
    plt.imshow(img)
plt.show()
```
